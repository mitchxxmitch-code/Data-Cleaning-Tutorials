## Excourse Data Structures

```python
peoples_dict = {
    "First_Name": ["Max", "John", "Anna"],
    "Last_Name": ["Mason", "Doe", "Adams"],
    "Email": ["max_mason939sd39@gmail.com", "john_doe939ds39@gmail.com", "anna_adams284s72d8@gmail.com"]
}
```

```python
peoples_df = pd.DataFrame.from_dict(peoples_dict)
peoples_df
```
