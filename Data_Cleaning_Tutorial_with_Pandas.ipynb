## Part 5

### Conditional Filtering 

```python
survey_df.head()
```

```python
python_devs = survey_df[survey_df["LanguageHaveWorkedWith"] == "Python"]
python_devs
```

```python
all_python_dev_filter = survey_df["LanguageHaveWorkedWith"].str.contains("Python", na=False)
all_python_devs = survey_df[all_python_dev_filter]
all_python_devs
```

```python
all_python_dev_filter = survey_df["LanguageHaveWorkedWith"].str.contains("Python", na=False)
brazil_filter = survey_df["Country"] == "Brazil"

all_brazil_python_devs = survey_df[all_python_dev_filter & brazil_filter]
all_brazil_python_devs
```
